### 3.1.1 CUSTOMER SEGMENTATION

`1.1 Bagaimana segmentasi pelanggan supermarket dalam menentukan pelanggan potensial untuk dijadikan target dalam meningkatkan penjualan?`

Dalam menentukan segmentasi pelanggan, dilakukan riset menggunakan metode RFM, yaitu membagi pelanggan menjadi beberapa segmen berdasarkan nilai *Recency*, *Frequency*, dan *Monetary*.

```python
#Membuat variabel dataframe RFM
df_RFM = df[['Recency', 'Frequency', 'Monetary']]
```

```python
# Membuat segmentasi untuk masing-masing fitur RFM
df_RFM['R_Segmen'] = np.where(df_RFM['Recency'] > df_RFM['Recency'].median(), 0, 1)
df_RFM['F_Segmen'] = np.where(df_RFM['Frequency'] > df_RFM['Frequency'].median(), 1, 0)
df_RFM['M_Segmen'] = np.where(df_RFM['Monetary'] > df_RFM['Monetary'].median(), 1, 0)
```

```python
# Membuat fitur RFM_Score
df_RFM['RFM_Score'] = df_RFM['R_Segmen'].astype(str) + df_RFM['F_Segmen'].astype(str) + df_RFM['M_Segmen'].astype(str)
```

```python
# Membuat segmentasi berdasarkan seluruh fitur RFM
df_RFM['RFM_Segmen'] = np.where((df_RFM['R_Segmen']==1) & (df_RFM['F_Segmen']==1) & (df_RFM['M_Segmen']==1), 1,
                        np.where((df_RFM['R_Segmen']==0) & (df_RFM['F_Segmen']==1) & (df_RFM['M_Segmen']==1), 1,
                        np.where((df_RFM['R_Segmen']==1) & (df_RFM['F_Segmen']==0) & (df_RFM['M_Segmen']==1), 1,
                        np.where((df_RFM['R_Segmen']==1) & (df_RFM['F_Segmen']==1) & (df_RFM['M_Segmen']==0), 2,
                        np.where((df_RFM['R_Segmen']==0) & (df_RFM['F_Segmen']==0) & (df_RFM['M_Segmen']==1), 1,
                        np.where((df_RFM['R_Segmen']==0) & (df_RFM['F_Segmen']==1) & (df_RFM['M_Segmen']==0), 2,
                        np.where((df_RFM['R_Segmen']==1) & (df_RFM['F_Segmen']==0) & (df_RFM['M_Segmen']==0), 2, 3
                        )))))))

df_RFM
```

```python
# Melihat distribusi segmen pelanggan berdasarkan fitur recency, frequency, dan monetary

from mpl_toolkits.mplot3d import Axes3D

# Create a figure and axis
fig = plt.figure(figsize=(15, 15))
ax = fig.add_subplot(projection='3d')

# Scatter plot with hue
scatter = ax.scatter(df['Recency'], df['Frequency'], df['Monetary'], c=df_RFM['RFM_Segmen'], cmap='viridis')

# Set labels
ax.set_xlabel('Recency')
ax.set_ylabel('Frequency')
ax.set_zlabel('Monetary')

# Add legend
legend1 = ax.legend(*scatter.legend_elements(), title='RFM_Segmen')
ax.add_artist(legend1)

# Set title
plt.title('Customer Distribution Based on RFM Features')

# Show plot
plt.show()
```

```python
# Melihat jumlah pelanggan pada setiap segmen RFM

# Get the count of each RFM segment
segment_counts = df_RFM['RFM_Segmen'].value_counts()

# Sort segments by count
ordered_segments = segment_counts.index

# Create the countplot with custom palette and sorted segments
plt.figure(figsize=(15, 6))
ax = sns.countplot(x='RFM_Segmen', data=df_RFM)
plt.title('Jumlah Pelanggan dalam Setiap RFM Segment')
plt.xlabel('RFM Segment')
plt.ylabel('Jumlah Pelanggan')

# Menambahkan label pada masing-masing bar
for p in ax.patches:
    ax.text(p.get_x() + p.get_width() / 2., p.get_height(), f'{int(p.get_height())}', ha='center', va='bottom', fontsize=10)

plt.show()
```

```python
# Menghitung jumlah pelanggan pada setiap segmen RFM
segment_counts = df_RFM['RFM_Segmen'].value_counts()

# Membuat DataFrame dari jumlah pelanggan
df_segment_counts = pd.DataFrame(segment_counts)
df_segment_counts.columns = ['Count']

# Mengurutkan DataFrame berdasarkan segmen RFM
df_segment_counts = df_segment_counts.sort_index()

# Menambahkan baris total
df_segment_counts.loc['Total'] = df_segment_counts['Count'].sum()

df_segment_counts = df_segment_counts.T
df_segment_counts
```

```python
# Menggabungkan variabel dataframe utama dengan dataframe RFM
df_concat_RFM = pd.concat([df, df_RFM['RFM_Segmen']], axis=1)
```

Segmen pelanggan dengan nilai *monetary* di atas rata-rata, yaitu pelanggan yang cenderung melakukan transaksi dengan nominal yang besar. Terdiri dari pelanggan dengan RFM_Score 111, 011, 101, 001. <br><br>
2. `Segmen 2`<br>
Segmen pelanggan dengan nilai *frequency* dan/atau *recency* di atas rata-rata, yaitu pelanggan yang cenderung sering melakukan transaksi dan/atau baru melakukan transaksi dalam waktu dekat. Terdiri dari pelanggan dengan RFM_Score 110, 010, dan 100.<br><br>
3. `Segmen 3`<br>
Segmen pelanggan dengan nilai *recency*, *frequency*, dan *monetary* di bawah rata-rata, yaitu pelanggan yang cenderung sudah lama tidak melakukan transaksi, jarang melakukan transaksi, dan dengan nominal transaksi yang kecil. Terdiri dari pelanggan dengan RFM_Score 000.

**RELATIONSHIP BETWEEN RFM SEGMEN AND CUSTOMER DEMOGRAPHIC PROFILE**

Dalam menentukan hubungan antara segmen pelanggan dengan profil demografi, dilakukan teknik **Bivariate Analysis** untuk mengeksplorasi adakah hubungan atau tren antara segmen pelanggan dengan profil usia, pendidikan, status pasangan, jumlah anak, dan jumlah pendapatan.

```python
# Melihat hubungan profil demografi dengan segmen pelanggan potensial

# Membuat kelompok/bin usia
age_bins = [0, 36, 51, 66, float('inf')]
age_labels = ['0-35', '36-50', '51-65', '66+']
df_concat_RFM['Age_Group'] = pd.cut(df_concat_RFM['Age'], bins=age_bins, labels=age_labels, right=False)

# Membuat kelompok/bin pendapatan
income_bins = [0, 40001, 80001, float('inf')]
income_labels = ['0-40000', '40001-80000', '>80001']
df_concat_RFM['Income_Group'] = pd.cut(df_concat_RFM['Income'], bins=income_bins, labels=income_labels, right=False)

# Menggabungkan 'Bachelor' dan 'Master' menjadi satu kategori 'Bachelor/Master'
df_concat_RFM['Education_Group'] = df_concat_RFM['Education'].replace({'Bachelor': 'Bachelor/Master', 'Master': 'Bachelor/Master'})

# Mengatur urutan kategori untuk kolom Education
education_order = ['Basic', 'Bachelor/Master', 'PhD']
df_concat_RFM['Education_Group'] = pd.Categorical(df_concat_RFM['Education_Group'], categories=education_order, ordered=True)

# Kolom kategorikal yang ingin ditampilkan
column_categoric = ['Age_Group', 'Education_Group', 'Partner_Status', 'Num_Child', 'Income_Group']

# Jumlah barplot
num_plots = len(column_categoric)

# Ukuran keseluruhan subplot
plt.figure(figsize=(15, 10))

for i in range(num_plots):
    column = column_categoric[i]
    # Membuat variabel df untuk kolom kategorikal
    df_bivariate = df_concat_RFM[[column, 'RFM_Segmen']]

    # Jika kolom merupakan kolom dengan bining (Income_Group atau Age_Group)
    if column in ['Income_Group', 'Age_Group']:
        # Melihat rata-rata RFM segmen dari grup yang telah dibin
        bivariate_segmen = df_bivariate.groupby([column]).mean().sort_index(ascending=True)
        # Mendapatkan label kolom yang sesuai
        column_label = column
    else:
        # Melihat rata-rata RFM segmen dari kolom kategorikal
        bivariate_segmen = df_bivariate.groupby([column]).mean().sort_index(ascending=True)
        column_label = column

    # Menentukan subplot
    plt.subplot(2, 3, i + 1)

    # Plot barplot
    sns.barplot(x=bivariate_segmen.index, y='RFM_Segmen', data=bivariate_segmen.reset_index(), palette='viridis')
    plt.xlabel(column_label)
    plt.ylabel('RFM Segmen Average')
    plt.title(f'RFM Segmen Average Based on {column_label}')

    # Menambahkan garis tren kecuali untuk "Parent_Status" dan "Marital_Status"
    if column not in ['Parent_Status', 'Marital_Status']:
        sns.lineplot(x=bivariate_segmen.index, y=bivariate_segmen['RFM_Segmen'], color='red', marker='o', label='Trend Line')

# Menambahkan legend untuk keseluruhan plot
plt.legend(title='Legend', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.show()
```

**TEMUAN**

Berdasarkan hasil analisis menggunakan metode bivariat, didapat informasi mengenai hubungan antara segmen pelanggan dengan profil demografi, diantaranya:

1. `Usia` <br>
Semakin tua usia pelanggan maka semakin tinggi kecenderungan pelanggan untuk sering melakukan transaksi dengan nominal yang besar dan baru dilakukan dalam waktu dekat.<br><br>

3. `Status Pasangan` <br>
Tidak ada hubungan yang signifikan antara segmen pelanggan dengan yang memiliki pasangan hidup dengan yang tidak memiliki pasangan hidup<br><br>

4. `Jumlah Anak` <br>
Semakin banyak jumlah anak yang dimiliki pelanggan maka semakin rendah kecenderungan pelanggan untuk sering melakukan transaksi dengan nominal yang besar dan baru dilakukan dalam waktu dekat<br><br>

**DISTRIBUTION CHANNEL PREFERENCE**

Dalam menentukan *distribution channel* yang cocok untuk melayani segmen pelanggan potensial, dilakukan analisa data dengan membandingkan distribusi pelanggan dalam masing-masing *channel*.

```python
# Membuat variabel dataframe *distribution channel*
df_distribution = df_concat_RFM[['NumWebPurchases', 'NumStorePurchases', 'NumCatalogPurchases', 'RFM_Segmen']]

## Membuat tabel jumlah pembelian pada masing-masing *distribution channel* setiap segmen pelanggan
distribution_segmen = df_distribution.groupby('RFM_Segmen').sum()
distribution_segmen.loc['Total'] = distribution_segmen.sum()
distribution_segmen
```

```python
# Mengubah tampilan tabel dengan Transpose
distribution_segmen = distribution_segmen.T
distribution_segmen['Total'] = distribution_segmen.sum(axis=1)
distribution_segmen['Total']=distribution_segmen.sum(axis=1)
distribution_segmen = distribution_segmen[['Total',1,2,3]]

# Menampilkan data dalam bentuk persentase
distribution_segmen = distribution_segmen.sort_values(by='Total', ascending=False)
distribution_segmen = distribution_segmen.div(distribution_segmen.sum(),axis=1)*100
distribution_segmen.fillna(0).round(2)
```

```python
# Menyimpan dataframe dalam bentuk csv
distribution_csv = distribution_segmen.copy()
distribution_csv.reset_index(inplace=True)
pd.melt(distribution_csv, id_vars=['index'], var_name='Segmen_Distribution', value_name='Percentage_Distribution').to_csv('distribution.csv', index=False)
```
